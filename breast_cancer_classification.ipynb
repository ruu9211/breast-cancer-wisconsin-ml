### Import the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import FunctionTransformer, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
import pickle
import warnings
warnings.filterwarnings('ignore')
```

#### Dataset Summary and Outlier Analysis

The dataset consists of 569 observations with multiple numeric features describing breast cancer tumor characteristics. Two columns should be removed during preprocessing: Unnamed: 32, which contains only null values, and id, which is simply an identifier and has no relationship with the target variable, diagnosis. The target variable, diagnosis, is categorical, representing benign and malignant cases, and will need to be converted into numerical values prior to modeling. Reviewing the descriptive statistics, several features show a large jump between the 75th percentile and the maximum value, suggesting potential outliers or rare extreme cases. Specifically, these include **area_mean (782.7 → 2501.0)**, **perimeter_worst (125.4 → 251.2)**, **area_worst (1084.0 → 4254.0)**, **compactness_worst (0.339 → 1.058**), **concavity_worst (0.383 → 1.252)**, **concave points_worst (0.161 → 0.291)**, **fractal_dimension_worst (0.092 → 0.2075)**, **concave_points_mean (0.074 → 0.2012)**, and **symmetry_worst (0.318 → 0.664)**. These large gaps may indicate the presence of outliers due to measurement error or reflect rare but valid extreme tumor cases. Further investigation is necessary to determine whether these values should be removed or retained for modeling.

```python
# removing null values
df = df.drop('Unnamed: 32', axis=1)
```

```python
# removing ID column from the dataset
df = df.drop('id', axis=1)
```

```python
# convert diagnosis column to numerical values
df['diagnosis'].unique()
```

As seen above, the diagnosis descriptions are as follows: Malignant and Benign.

The above descriptions are converted to numerical values: 0 --> Benign, 1 --> Malignant

```python
df['diagnosis'] = df['diagnosis'].map({'B':0, 'M':1})
```

### Data Preparation and Pipeline Setup

```python
# assign X and y variables
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
# split the dataset for training and testing
# assign stratify=y since we know that our dataset is imbalanced
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, stratify=y, random_state=42)

# check split
print(X_train.shape)
print(X_test.shape)
```

```python
# define function for log-transformation to normalize the dataset
def log_transformation(X):
    X_log = np.log1p(X)
    return X_log

# wrap the function inside a FunctionTransformer
log_transform_object = FunctionTransformer(log_transformation)

# features which needs to undergo log-transformation
log_cols = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean',
            'concave points_mean', 'fractal_dimension_mean','radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
            'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se','fractal_dimension_se', 'radius_worst','perimeter_worst', 
            'area_worst', 'compactness_worst', 'concavity_worst', 'symmetry_worst', 'fractal_dimension_worst']
    
# apply function to select features
log_transform = ColumnTransformer([('log_transform', log_transform_object, log_cols)], remainder='passthrough')
```

### Model Experimentation 🧪 

#### 1) Logistic Regression Model ➗📈

```python
# build the pipeline for Logistic Regression
lr_pipe = Pipeline([('log_transform', log_transform),
                    ('scaler', RobustScaler()),
                    ('select_best', SelectKBest()),
                    ('classifier', LogisticRegression()) ])

# assign param_grid
param_grid = {'select_best__k': [5, 10, 15, 20, 25]}

# perform cross validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# perform GridSearchCV to optimize our model and find the best parameters
lr_grid_search = GridSearchCV(estimator=lr_pipe, param_grid=param_grid, cv=cv, n_jobs=-1)

# fit the model on training set
lr_grid_search.fit(X_train, y_train)
                                   
# predict the model on testing set
y_pred = lr_grid_search.predict(X_test)

# print best CV score and parameters
print(f'Best CV Score: {lr_grid_search.best_score_*100:.2f}%')
print(f'Best Parameters: {lr_grid_search.best_params_}')
```

```python
# perform metric evaluations to determine model performance
train_lr_accuracy = accuracy_score(y_train, lr_grid_search.predict(X_train))
lr_accuracy = accuracy_score(y_test, y_pred)
lr_classification = classification_report(y_test, y_pred)
lr_cm = confusion_matrix(y_test, y_pred)

print(f'Logistic Regression Model Accuracy (Train Set): {train_lr_accuracy*100:.2f}%\n')
print(f'Logistic Regression Model Accuracy (Test Set): {lr_accuracy*100:.2f}%\n')
print(f'Logistic Regression Model Classification Report: \n{lr_classification}\n')

# plot confusion matrix for analysis purposes
plt.figure(figsize=(6,4))
ax = sns.heatmap(lr_cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign (Pred)', 'Malignant (Pred)'], yticklabels=['Benign (Actual)', 'Malignant (Actual)'])
ax.set_title('Confusion Matrix')
ax.set_ylabel('Actual')
ax.set_xlabel('Predicted')
```

#### 2) Random Forest Classifier Model 🌲🌳

```python
# build the pipeline for Random Forest Classifier
rfc_pipe = Pipeline([('scaler', RobustScaler()),
                     ('select_best', SelectKBest()),
                     ('classifier', RandomForestClassifier(random_state=42)) ])

# assign param_grid
param_grid = {'select_best__k': [5, 10, 15, 20, 25],
              'classifier__n_estimators': np.arange(100, 501, 100),
              'classifier__criterion': ['gini', 'entropy'],
              'classifier__max_depth': [None, 5, 10, 15, 20],
              'classifier__min_samples_split': [2, 5, 10]
             }
              
# perform cross validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# perform GridSearchCV to optimize our model and find the best parameters
rfc_grid_search = GridSearchCV(estimator=rfc_pipe, param_grid=param_grid, cv=cv, n_jobs=-1)

# fit the model on training set
rfc_grid_search.fit(X_train, y_train)
                                   
# predict the model on testing set
y_pred = rfc_grid_search.predict(X_test)

# print best CV score and parameters
print(f'Best CV Score: {rfc_grid_search.best_score_*100:.2f}%')
print(f'Best Parameters: {rfc_grid_search.best_params_}')
```

```python
# perform metric evaluations to determine model performance
train_rfc_accuracy = accuracy_score(y_train, rfc_grid_search.predict(X_train))
rfc_accuracy = accuracy_score(y_test, y_pred)
rfc_classification = classification_report(y_test, y_pred)
rfc_cm = confusion_matrix(y_test, y_pred)

print(f'Random Forest Classifier Model Accuracy (Train Set): {train_rfc_accuracy*100:.2f}%\n')
print(f'Random Forest Classifier Model Accuracy (Test Set): {rfc_accuracy*100:.2f}%\n')
print(f'Random Forest Classifier Classification Report: \n{rfc_classification}\n')

# plot confusion matrix for analysis purposes
plt.figure(figsize=(6,4))
ax = sns.heatmap(rfc_cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign (Pred)', 'Malignant (Pred)'], yticklabels=['Benign (Actual)', 'Malignant (Actual)'])
ax.set_title('Confusion Matrix')
ax.set_ylabel('Actual')
ax.set_xlabel('Predicted')
```

### Model Comparison and Selection ⚖️

**See below table for summary of results**


| 🔍 Metric              | 📈 LR  | 🌲 RFC |
|------------------------|--------|--------|
| 🎯 Train Accuracy      | 98.46% | 99.78% |
| 🎯 Test Accuracy       | 98.25% | 97.37% |
| ✅ Precision (Class 0) | 0.99   | 0.96   |
| 🔄 Recall (Class 0)    | 0.99   | 1.00   |
| ✅ Precision (Class 1) | 0.98   | 1.00   |
| 🔄 Recall (Class 1)    | 0.98   | 0.93   |
| 🧪 Best CV Score       | 97.36% | 97.14% |


✅ The final model selected is **Logistic Regression** due to the following reasons:

- **Generalization:** Logistic Regression shows consistent performance between training (98.46%) and testing (98.25%) sets, while Random Forest shows stronger training accuracy (99.78%) but lower test accuracy (97.37%).
- **Recall on Positive Class (Malignant):** Logistic Regression achieves higher recall (0.98 vs 0.93), meaning it detects more malignant tumors correctly — a crucial factor in medical applications.
- **Precision Tradeoff:** Random Forest has slightly better precision for malignant tumors (1.00 vs 0.98), but at the cost of recall, which could result in missed detections.

👉 For these reasons, **Logistic Regression** is the preferred model. It balances precision and recall more effectively and generalizes better.
